```python
# ── Challenge 1: The Data Profiler ──────────────────────────

# Step 1: Our raw data — strings with extra spaces and weird casing
raw_readings = [" 23.5 ", "low", " 45.0", "HIGH ", "30.2"]

# Step 2: Loop through, clean up each item
cleaned_readings = []          # empty list we'll fill up

for item in raw_readings:
    cleaned = item.strip()     # .strip() removes spaces from both ends
    cleaned = cleaned.lower()  # .lower() makes EVERYTHING lowercase
    cleaned_readings.append(cleaned)  # add to our cleaned list

print("Cleaned:", cleaned_readings)


# Step 3: Try to convert each item to a float (decimal number)
numeric_readings = []          # empty list for numbers only

for item in cleaned_readings:
    try:
        number = float(item)       # try converting to float
        numeric_readings.append(number)  # if it works, save it
    except ValueError:
        pass                       # "low", "high" will fail → we skip them

print("Numeric only:", numeric_readings)
```

```python
# ── Challenge 2: The Trend Analyzer ─────────────────────────

def analyze_stock_prices(prices):
    """
    prices = a dictionary like {"2024-01-01": 150.0, "2024-01-02": 162.0}
    """

    # Step 1: Calculate the average price using a for loop
    total = 0
    for date in prices:
        total = total + prices[date]   # add each day's price to total

    average = total / len(prices)      # len() gives us the count of days
    print(f"Average price: {average:.2f}")

    # Step 2: Check each day against the average
    results = []                       # will hold our (date, status) tuples

    for date in prices:
        price = prices[date]           # get that day's price

        if price > average * 1.05:     # more than 5% above average?
            status = "High Volatility"
        elif price < average * 0.95:   # more than 5% below average?
            status = "Low Volatility"
        else:
            status = "Stable"

        results.append((date, status)) # store as a tuple (pair)

    return results
# ── Test it out ──────────────────────────────────────────────
stock_prices = {
    "2024-01-01": 150.0,
    "2024-01-02": 162.0,   # spike!
    "2024-01-03": 148.0,
    "2024-01-04": 135.0,   # dip!
    "2024-01-05": 152.0,
}

output = analyze_stock_prices(stock_prices)

for date, status in output:
    print(f"{date}: {status}")
```
