```python
# ── Challenge 1: The Data Profiler ──────────────────────────

# Step 1: Our raw data — strings with extra spaces and weird casing
raw_readings = [" 23.5 ", "low", " 45.0", "HIGH ", "30.2"]

# Step 2: Loop through, clean up each item
cleaned_readings = []          # empty list we'll fill up

for item in raw_readings:
    cleaned = item.strip()     # .strip() removes spaces from both ends
    cleaned = cleaned.lower()  # .lower() makes EVERYTHING lowercase
    cleaned_readings.append(cleaned)  # add to our cleaned list

print("Cleaned:", cleaned_readings)


# Step 3: Try to convert each item to a float (decimal number)
numeric_readings = []          # empty list for numbers only

for item in cleaned_readings:
    try:
        number = float(item)       # try converting to float
        numeric_readings.append(number)  # if it works, save it
    except ValueError:
        pass                       # "low", "high" will fail → we skip them

print("Numeric only:", numeric_readings)
```
