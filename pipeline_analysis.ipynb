```python
# ── Challenge 1: The Data Profiler ──────────────────────────


raw_readings = [" 23.5 ", "low", " 45.0", "HIGH ", "30.2"]


cleaned_readings = []          

for item in raw_readings:
    cleaned = item.strip()     
    cleaned = cleaned.lower()  
    cleaned_readings.append(cleaned)  

print("Cleaned:", cleaned_readings)



numeric_readings = []          

for item in cleaned_readings:
    try:
        number = float(item)       
        numeric_readings.append(number) 
    except ValueError:
        pass                      
print("Numeric only:", numeric_readings)
```

```python
# ── Challenge 2: The Trend Analyzer ─────────────────────────

def analyze_stock_prices(prices):
    """
    prices = a dictionary like {"2024-01-01": 150.0, "2024-01-02": 162.0}
    """

    total = 0
    for date in prices:
        total = total + prices[date]   

    average = total / len(prices)     
    print(f"Average price: {average:.2f}")

    
    results = []                       

    for date in prices:
        price = prices[date]          

        if price > average * 1.05:    
            status = "High Volatility"
        elif price < average * 0.95:   
            status = "Low Volatility"
        else:
            status = "Stable"

        results.append((date, status)) 

    return results

stock_prices = {
    "2024-01-01": 150.0,
    "2024-01-02": 162.0,   
    "2024-01-03": 148.0,
    "2024-01-04": 135.0,   
    "2024-01-05": 152.0,
}

output = analyze_stock_prices(stock_prices)

for date, status in output:
    print(f"{date}: {status}")
```
